<h2> Understand about Dataset</h2>
This dataset contains information about loan applicants and their financial background. It includes personal details, income, assets, credit score, and loan related attributes, along with the final decision on whether the loan was approved or rejected. Understanding these features is important because each of them plays a role in the loan approval process

```python
df.columns
```


<h2>Below is a brief explanation of each feature in the dataset:</h2>

**loan_id:** A unique identification number assigned to each loan application

**no_of_dependents:** Number of dependents the applicant is financially responsible for. A higher number of dependents may affect repayment capacity

**education:** Educational qualification of the applicant (Graduate / Not Graduate), which can influence income stability

**self_employed:** Indicates whether the applicant is self-employed or not (Yes / No)

**income_annum:** Annual income of the applicant. Higher income generally increases the chances of loan approval

**loan_amount:** The total loan amount requested by the applicant

**loan_term:** Duration (in years) for which the loan is taken. Longer terms reduce EMI but increase risk

**cibil_score:** Credit score of the applicant, reflecting past credit behavior. Higher scores indicate better creditworthiness

**residential_assets_value:** Total value of residential assets owned by the applicant

**commercial_assets_value:** Total value of commercial assets owned by the applicant

**luxury_assets_value:** Value of luxury assets such as vehicles, jewelry, or high-end properties

**bank_asset_value:** Amount of assets or savings available in bank accounts

**loan_status:** Target variable indicating whether the loan was Approved or Rejected

<h4>making the data balanced</h4> by applying log transformation to numerical features like income, loan amount, and asset values.

**Why you are doing this**

These columns have very large and skewed values

Log transformation reduces skewness, making the data more normally distributed.

t controls the outliers.

Helps models learn patterns more accurately and stably.

```python
x = df.drop('loan_status', axis=1)
 #This line separates the input features by removing the target variable loan_status from the dataset before model training
```

```python
y = df['loan_status']
#This line assigns the target variable  loan_status to y, which the model will learn to predict
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)   # Approved=1, Rejected=0
```

converts the categorical target variable loan_status into numerical labels, where

Approved = 1

Rejected = 0

```python
x = pd.get_dummies(x, drop_first=True)
```

This line changes text categories into numbers (0 and 1) so the machine learning model can understand them, and it removes one extra column to prevent duplicate or confusing information during training.

**for example**
<h3>Before</h3>
education: Graduate, Not Graduate

<h3>After</h3>
education
1 = Graduate &
0 = Not Graduate



```python
import numpy as np

x = x.replace([np.inf, -np.inf], np.nan)
```

This line finds very large or very small infinite values in the data and replaces them with empty values, so the model doesn’t crash or get confused while learning.

```python
x = x.fillna(0)
#this fills missing values with 0
```

Installing imblearn library for Handling Imbalanced Data 

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors=4) #SMOTE = Synthetic Minority Oversampling Technique
x,y = oversample.fit_resample(x.fillna(0), y)
```

Earlier we saw imbalance in target class now,

This balances the target classes means our loan_status using SMOTE by creating synthetic samples of the minority class, so the model gets an equal number of Approved and Rejected loans for better learning.


```python
import pandas as pd
pd.Series(y).value_counts() #this shows how many samples belong to each class
```

This output shows that after applying SMOTE, the dataset is now balanced:

0 → 2656 samples (Rejected loans)

1 → 2656 samples (Approved loans)

```python
x.describe() #it show the statical data
```

**scales all feature values between 0 and 1 using Min-Max scaling**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
x = pd.DataFrame(x)
x.head()
```

Min-Max Scaling to normalize all features in the dataset to a 0–1 range. 

Features like income or loan amount have very large values compared to others, which can dominate the learning process.

Scaling ensures that all features contribute proportionally

**splitting the dataset into training and testing sets**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

This divides the dataset into training and testing sets
 keeping 
 
 75% of the data for training the model 

 25% for testing its performance
 
  It ensures that the model learns patterns from one set and is evaluated on unseen data to check its accuracy and generalization

<h3>Loading the LogisticRegression model</h3>

```python
from sklearn.linear_model import LogisticRegression
model_LogReg = LogisticRegression()
model_LogReg.fit(x_train,y_train)
```

This calculates the accuracy of the Logistic Regression model on the test data and rounds it to a percentage, showing how well the model predicts loan approval

```python
#here it shows 92% Accuracy
from sklearn.metrics import accuracy_score 
LogReg_acc=round(model_LogReg.score(x_test,y_test)*100)
LogReg_acc
```

<h3>Loading the Support Vector Machine model</h3>

```python
from sklearn.svm import SVC
model_SVM = SVC(kernel="linear")
model_SVM.fit(x_train,y_train)
```

This code calculates the accuracy of the supportVectorMachine model on the test data and rounds it to a percentage, showing how well the model predicts loan approval

```python
SVM_acc=round(model_SVM.score(x_test,y_test)*100) #here it shows 93% Accuracy
SVM_acc
```

<h3>Loading the DecisionTreeClassifier model</h3>

```python
from sklearn.tree import DecisionTreeClassifier
model_DT = DecisionTreeClassifier()
model_DT.fit(x_train,y_train)
```

This code calculates the accuracy of the DesicionTreeClassifier model on the test data and rounds it to a percentage, showing how well the model predicts loan approval

```python
DT_acc=round(model_DT.score(x_test,y_test)*100) ##here it shows 98% Accuracy
DT_acc
```

<h3>Loading the RandomForestClassifier model</h3>

```python
from sklearn.ensemble import RandomForestClassifier 
model_RF = RandomForestClassifier()
model_RF.fit(x_train,y_train)
```

This code calculates the accuracy of the RandomForestClassifier model on the test data and rounds it to a percentage, showing how well the model predicts loan approval

```python
RF_acc=round(model_RF.score(x_test, y_test)*100)
RF_acc 
#here it is also shows 98% Accuracy
```

Lets compare and visualize the performance of multiple models on the same dataset

```python
Accuracy = [LogReg_acc,DT_acc,RF_acc,SVM_acc]
models = ['LogisticRegression', 'DecisionTreeClassifier' , 'RandomForestClassifier', 'Support Vector Machine']
```

**Model Accuracy Comparison**

```python
sns.barplot(x = Accuracy, y = models, palette = 'hls')
plt.xlabel("Accuracy in %")
plt.title("Accuracy")
plt.show()
```

This plot shows a comparison of accuracies for different models, helping to identify which model performs best on the dataset.

<h3><center>Confusion Matrix for Random Forest</center></h3>

```python
from sklearn.metrics import confusion_matrix
y_pred = model_RF.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
classes = model_RF.classes_
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes,yticklabels=classes)
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

This creates a confusion matrix and visualizes it as a heatmap for the Random Forest model:

Shows true positives, true negatives, false positives, and false negatives.

Helps evaluate how well the model predicts Approved (1) and Rejected (0) loans.

The heatmap makes it easy to see where the model is performing well or making mistakes.

<h3>To predict the loan approval status for a new applicant</h3>

```python

features = ['loan_id', 'no_of_dependents', 'income_annum', 'loan_amount', 'loan_term',
    'cibil_score', 'residential_assets_value', 'commercial_assets_value',
    'luxury_assets_value', 'bank_asset_value',
    'education_ Not Graduate', 'self_employed_ Yes']

feature_values = {
    'loan_id': 0,
    'no_of_dependents': 2,
    'income_annum': np.log(5000000),
    'loan_amount': np.log(15000000),
    'loan_term': 15,
    'cibil_score': 750,
    'residential_assets_value': np.log(8000000),
    'commercial_assets_value': np.log(3000000),
    'luxury_assets_value': np.log(10000000),
    'bank_asset_value': np.log(4000000),
    'education_ Not Graduate': 0,
    'self_employed_ Yes': 0
}

user_input = pd.DataFrame([feature_values], columns=features)
user_input = scaler.transform(user_input)
prediction = model_RF.predict(user_input)

print("predicted loan status:", prediction)
```

**Input features:** Values for the applicant’s dependents, income, loan amount, loan term, CIBIL score, assets, education, and employment status.

**Preprocessing:** Applies log transformation to skewed numerical features and scales all features using the same scaler used during model training.

**Prediction:** The preprocessed input is fed into the trained Random Forest model, which outputs 1 for Approved or 0 for Rejected loans.

**Output:** Displays the predicted loan status, which can be converted to a human readable form like “Approved” or “Rejected”
